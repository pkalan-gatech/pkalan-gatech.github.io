```python
#@title Imports
import matplotlib.pyplot as plt
import numpy as np
# from scipy import interpolate
from scipy.interpolate import make_interp_spline
import pandas as pd
```

```python
#@title Salary dataset
# data from levels.fyi
positions = ["Mechanical Engineer", "Software Engineer", "Data Scientist", "Product Manager", "Program Manager", "Project Manager", "Technical Program Manager", "ML/AI SE"]
percentiles = np.array([25, 50, 75, 90])
salaries = np.array([
[113,165,150,180,160,125,191, 172],
[142,220,200,247,202,160,255, 246],
[188,300,266,340,255,213,336, 348],
[252,400,344,470,319,275,420, 450]
])
data = {'Mechanical Engineer':[113, 142, 188, 252],
        'Software Engineer': [165, 220, 300, 400],
        'Data Scientist': [150, 200, 266, 344],
        'ML/AI Software Engineer': [172, 246, 348, 450],
        'Technical Program Manager': [191, 255, 336, 420]
        }
df = pd.DataFrame(data, index = [25, 50, 75, 90])
```

```python
#@title Comparison of job positions by salaries (for the state of California):
df.plot(style='.-', xlabel='Percentile', ylabel='Salary in $k', legend=False, figsize=(5,3), grid=True);
for entry in data:
  plt.text(94, data[entry][-1]-5, entry)
```

```python
#@title Comparison of other jobs with mechanical engineering in terms of salary (for the state of California):
df_norm = df.div(df['Mechanical Engineer'], axis=0)
df_norm.plot(style='.-', xlabel='Percentile', ylabel='Salary compared with Mech. Engg.', legend=False, figsize=(5,3), grid=True);
for entry in df_norm:
  plt.text(94, df_norm[entry][90]-0.02, entry)
```
